```python
def LimpiezaPuntos(df,var_name):
    df[var_name] = df[var_name].astype('O')
    df[var_name] = df[var_name].replace(".","").astype('str')
    return df

def LimpiezaCaracteres(df,var_name):
    df[var_name] = df[var_name].astype('str')
    df[var_name] = df[var_name].str.replace("-","")
    df[var_name] = df[var_name].str.replace(" ","")
    df[var_name] = df[var_name].str.replace("?","")
    df[var_name] = df[var_name].str.replace("°","")
    df[var_name] = df[var_name].str.replace("(","")
    df[var_name] = df[var_name].str.replace(")","")
    df[var_name] = df[var_name].str.replace("()","")
    df[var_name] = df[var_name].str.replace(",","")
    df[var_name] = df[var_name].str.replace("¿","")
    df[var_name] = df[var_name].str.replace("|","")
    df[var_name] = df[var_name].str.replace("_","")
    df[var_name] = df[var_name].str.replace("@","")
    df[var_name] = df[var_name].str.replace("'","")
    df[var_name] = df[var_name].str.replace(":","")
    df[var_name] = df[var_name].str.replace("i","")
    df[var_name] = df[var_name].str.replace("ia","")
    df[var_name] = df[var_name].str.replace("q","")
    df[var_name] = df[var_name].str.replace(".","")
    df[var_name] = df[var_name].str.replace("PE","")
    df[var_name] = df[var_name].str.replace("PEP","")
    df[var_name] = df[var_name].str.replace("P","")
    df[var_name] = df[var_name].str.replace("v","")
    df[var_name] = df[var_name].str.replace("&","")
    df[var_name] = df[var_name].str.replace("Q","")
    df[var_name] = df[var_name].str.replace("O","")
    df[var_name] = df[var_name].str.replace("DI","")
    df[var_name] = df[var_name].str.replace("I","")
    df[var_name] = df[var_name].str.replace("D","")
    df[var_name] = df[var_name].str.replace("V","")
    df[var_name] = df[var_name].str.replace("Z","")
    df[var_name] = df[var_name].str.replace("a","")
    df[var_name] = df[var_name].str.replace("*","")
    df[var_name] = df[var_name].str.replace("pep","")
    return df

def DecimalLatitud(df,var_name):
    pd.options.mode.chained_assignment = None
    df[var_name] = df[var_name].replace("nan",0)
    df[var_name] = df[var_name].fillna(0)
    df['erroreslat'] = "no"
    z = len(df)
    for i in range(0,z):
        v = str(type(df[var_name][i]))
        if "float" in str(type(df[var_name][i])):
            df[var_name][i] = str(df[var_name][i])
            df[var_name][i] = df[var_name][i].replace(".","")
        if "E" in str(df[var_name][i]):
            df[var_name][i] = 0
            df['erroreslat'][i] = "si"
        if "." in str(df[var_name][i]):
            df[var_name][i] = (str(df[var_name][i]).replace(".",""))
        if "," in str(df[var_name][i]):
            df[var_name][i] = (str(df[var_name][i]).replace(",",""))
        if ":" in str(df[var_name][i]):
            df[var_name][i] = 0
            df['erroreslat'][i] = "si"
        if (df[var_name][i]==0):
            df['erroreslat'][i] = "si"
        if 'datetime.datetime' in v:
            df['erroreslat'][i] = "si"
        if not 'datetime.datetime' in v:
            if (len(str(df[var_name][i]))<=20):
                if (len(str(df[var_name][i]))==20):
                    b = int(df[var_name][i])*.0000000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==19):
                    b = int(df[var_name][i])*.000000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==18):
                    b = int(df[var_name][i])*.00000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                if (len(str(df[var_name][i]))==17):
                    b = int(df[var_name][i])*.0000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==16):
                    b = int(df[var_name][i])*.000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==15):
                    b = int(df[var_name][i])*.00000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==14):
                    b = int(df[var_name][i])*.0000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==13):
                    b = int(df[var_name][i])*.000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==12):
                    b = int(df[var_name][i])*.00000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==11):
                    b = int(df[var_name][i])*.0000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==10):
                    b = int(df[var_name][i])*.000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==9):
                    b = int(df[var_name][i])*.00000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==8):
                    b = int(df[var_name][i])*.0000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))==7):
                    b = int(df[var_name][i])*.000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d   
                if (len(str(df[var_name][i]))==6):
                    b = int(df[var_name][i])*.000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslat'][i] = "no"
                if (len(str(df[var_name][i]))<=5):
                    df['erroreslat'][i] = "si" 
                    
    return df

def DecimalLongitud(df,var_name):
    pd.options.mode.chained_assignment = None
    df[var_name] = df[var_name].replace("nan","0")
    df[var_name] = df[var_name].fillna(0)
    df['erroreslng'] = "no"
    z = len(df)
    for i in range(0,z):
        v = str(type(df[var_name][i]))
        if "float" in str(type(df[var_name][i])):
            df[var_name][i] = str(df[var_name][i])
            df[var_name][i] = df[var_name][i].replace(".","")
        if "E" in str(df[var_name][i]):
            df[var_name][i] = 0
            df['erroreslat'][i] = "si"
        if "." in str(df[var_name][i]):
            df[var_name][i] = (str(df[var_name][i]).replace(".",""))
        if "," in str(df[var_name][i]):
            df[var_name][i] = (str(df[var_name][i]).replace(",",""))
        if ":" in str(df[var_name][i]):
            df[var_name][i] = 0
            df['erroreslng'][i] = "si"
        if (df[var_name][i]==0):
            df['erroreslng'][i] = "si"
        if 'datetime.datetime' in v:
            df['erroreslng'][i] = "si"
        if not 'datetime.datetime' in v:
            if (len(str(df[var_name][i]))<=20):
                if (len(str(df[var_name][i]))==20):
                    b = int(df[var_name][i])*.00000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==19):
                    b = int(df[var_name][i])*.0000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==18):
                    b = int(df[var_name][i])*.000000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==17):
                    b = int(df[var_name][i])*.00000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==16):
                    b = int(df[var_name][i])*.0000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==15):
                    b = int(df[var_name][i])*.000000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==14):
                    b = int(df[var_name][i])*.00000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==13):
                    b = int(df[var_name][i])*.0000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==12):
                    b = int(df[var_name][i])*.000000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==11):
                    b = int(df[var_name][i])*.00000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==10):
                    b = int(df[var_name][i])*.0000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==9):
                    b = int(df[var_name][i])*.000001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==8):
                    b = int(df[var_name][i])*.00001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==7):
                    b = int(df[var_name][i])*.0001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))==6):
                    b = int(df[var_name][i])*.001
                    c = ("{0:.8f}".format(b))
                    d = float(c)
                    df[var_name][i] = d
                    df['erroreslng'][i] = "no"
                if (len(str(df[var_name][i]))<=5):
                    df['erroreslng'][i] = "si" 
                    
    return df

def LatitudFinal(df,var_name):
    LatInf=6.9
    LatSup=7.2
    df["LatFinal1"]="no"
    df["LatFinal2"]="no"
    z = len(df)
    for i in range(0,z):
        if (df[var_name][i]>=LatInf):
            df["LatFinal1"][i] = "si" 
        if (df[var_name][i]<=LatSup):
            df["LatFinal2"][i] = "si"
    return df

def LongitudFinal(df,var_name):
    LngInf=-73.00
    LngSup=-73.2
    df["LngFinal1"]="no"
    df["LngFinal2"]="no"
    z = len(df)
    for i in range(0,z):
        if (df[var_name][i]<=LngInf):
            df["LngFinal1"][i] = "si" 
        if (df[var_name][i]>=LngSup):
            df["LngFinal2"][i] = "si"
    return df

def ConcatConEspacio(df,var_name,var_name2,var_name3):
    df[var_name3] = df[var_name] +' '+ df[var_name2]
    return df
```
